```python
"""
Referencia: https://github.com/gallardorafael/edge2art
Este script convierte todas las imágenes del directorio en sus versiones
de bordes.
La transformación utiliza la implementacion de Canny Edge Detection de OpenCV
"""
```

```python
import cv2
import numpy as np
import os
from os.path import join, isfile, splitext, isdir
import argparse
```

```python
dir_input = "./tagged"
dir_output =  "./input"

def auto_canny(image, sigma=0.33):
    median = np.median(image)
    lower = int(max(0, (1.0 - sigma) * median))
    upper = int(min(255, (1.0 + sigma) * median))
    return cv2.Canny(image, lower, upper)

def preprocess(image_path):
    image = cv2.imread(image_path)
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    return cv2.GaussianBlur(gray_image, (3, 3), 0)

def draw_edges():
    count = 0
    # Se crean los directorios por estilo para las imágenes de bordes.
    if not isdir(dir_output):
        os.mkdir(dir_output)
    for artwork_name in os.listdir(dir_input):
        count += 1
        print("\n\nProcessing borders %d images\n"%(count))
        artwork_path = os.path.join(dir_input, artwork_name)
        image = preprocess(artwork_path)
        edge = 255 -  auto_canny(image)
        edge_three = cv2.merge([edge,edge,edge])
        target_path = os.path.join(dir_output, artwork_name)
        cv2.imwrite(target_path, edge_three)
```

```python
draw_edges()
```
